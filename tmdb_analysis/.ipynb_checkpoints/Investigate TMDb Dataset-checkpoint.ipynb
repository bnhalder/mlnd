### Data Cleaning

As shown above, there are a lot of null values in the columns. In the next few cells I'll perform some data cleaning steps to make the data useful for our future explorations. Please find the comments after each step. First I'll find out if there is any duplicate row in the dataset.

```python
# remove any duplicate row if exists
print(df.duplicated().sum())
df.drop_duplicates(inplace=True)
```

Found out that there is a duplicate row in the dataset. Removed it. Now, I'll examine null values.

```python
# check for null values
df.isnull().sum()
```

As examined above, a lot of null values are present in the dataset. A few rows have missing imdb_id and I plan to use imdb_id to uniquely idetify a row. So, I decided to remove those. I'll also remove rows which don't have any genre information as I plan to do data analysis based on genre. I'll not remove/process other columns which have null values as they will not be relevant for my analysis. All columns relevant to my analysis won't have null values after these steps.

```python
# drop rows which have null values in imdb_id and genres column
df.dropna(subset=['imdb_id'], inplace=True)
df.dropna(subset=['genres'], inplace=True)
df.info()
```

As shown above, now the dataset doesn't have null values in the fields of inerest. The next step will check if all the rows contain unique movies.

```python
# check identifier columns have any column has duplicated values
print(len(df['id'].unique()))
print(len(df['imdb_id'].unique()))
```

As seen in the result of the first cell, the genre column of the data frame is a list of concatenated strings which means one movie can belong to multiple genres. First, I'll find out full list of the genres for all the movies 

```python
# function to find out the full list of genres
def full_list_of_genres(genres):
    '''
    This function takes the genres column of the dataframe and 
    returns the full list of genres of all the movies in dataset
    '''
    genres_str = genres.values.tolist()
    genre_list = []
    for i in genres_str:
        split_genres = i.split('|')
        for j in split_genres:
            if j not in genre_list:
                genre_list.append(j)
    return genre_list

# Get the full list of genres
genres = full_list_of_genres(df['genres'])
```

Now, I'll create a new dataframe which will have  all the genre information against a particular movie. I'll go through the whole list of movies and find out the genres and put a 1 in the columns corresponding to the genre the movie belongs to. Then I'll merge the genre dataframe with the original dataframe on imdb_id and create a new dataframe which have the  consolidated information.

```python
# create a dataframe with genre info which can be indexed by imdb_id
genre_df = pd.DataFrame(index=range(len(df)) ,columns=genres)
genre_df = genre_df.fillna(value=0)

# fill up the genre df with genre info of each movie
for i in range(len(df)):
    genre_list = df.iloc[i]['genres'].split('|')
    genre_df.loc[i, 'imdb_id'] = df.iloc[i]['imdb_id']
    for j in genre_list:
        genre_df.loc[i, j] = 1
        
genre_df.head()
```

```python
# merge genre_df and original df to contain all the genre info as columns
tmdb_df = df.merge(genre_df, how='inner', on=['imdb_id'])

# save this cleaned dataframe as a csv, so that next time we can directly use this.
tmdb_df.to_csv('cleaned_data.csv', index=False)

#this is our clean dataframe on which we'll do all the data analysis
print(tmdb_df.shape)
tmdb_df.info()
```

Now the data is cleaned and contain new columns which corresponds to the genre the movie belong to. I'll use this new cleaned and engineered dataframe for analysis.

## Exploratory Data Analysis

### Invstigate the properties of the movies over each genre
This is to analyse how various properties of the movies varies over the genre. I'll investigate the distribution of average popularity, budget, runtime, vote-count, rating over all the genres. I'll also investigate how average popularity of each genre varies over time.   

```python
# Get the list of all genres
genres = full_list_of_genres(tmdb_df['genres'])

# Generic function to plot the distribution over genre
def distribution_over_genre(var):
    '''
    This function takes a variable name and plots the 
    distribution of that variable over genre on a bar chart
    '''
    distribution_var = pd.Series(index=genres)
    for genre in genres:
        distribution_var[genre] = tmdb_df.groupby([genre]).mean()[var][1]
    title = 'Distribution of '+ var +' over genre'
    distribution_var.plot(kind='bar', figsize=(20, 5), title=title)
    plt.show()
```

This is a generic utility code to plot the distribution of a particular field over genre. I'll use this below to plot the distribution as will be shown below.

```python
# The distribution of popularity over each genre
distribution_over_genre('popularity')
```

Here, the data indicates that Adventure, Science Fiction, Fantasy, Action, Animation, War movies tends to be more popular than other genres on average.

```python
# The distribution of budget over each genre
distribution_over_genre('budget')
```

Data indicates that Adventure, Fantasy, Action movies tend to have more budget than other genres.

```python
# The distribution of runtime over each genre
distribution_over_genre('runtime')
```

The data indicates that the average runtime doesn't vary much across genres. However, History, War, Western genre movies seems to be longer than others and Animation seems to have shorter runtime on average.

```python
# The distribution of vote count over each genre
distribution_over_genre('vote_count')
```

The data indicates that the following genres tend to the have the most votes - Adventure, Science Fiction, Fantasy, Action. This also suggests that people watches these genres more than the others. Foreign, Documentary and TV Movies are voted the least. 

```python
# The distribution of vote average(rating) over each genre
distribution_over_genre('vote_average')
```

The average rating however, seems to not vary much over the genres. And quite counterintuitively, Documentary, Animation movies get good ratings on average.

Next I'll investigate the variation in popularity over the years for various genres.

```python
# Now, I'll analyse how the popularity of each genre varies over time
years = sorted(tmdb_df['release_year'].unique(), reverse=True)
popularity_genre_year = pd.DataFrame(index=genres, columns=years)
popularity_genre_year = popularity_genre_year.fillna(value=0)
popularity_genre_year.head()
```

```python
# calculate average popularity of each genre on each year
for genre in genres:
    avg_popularity = tmdb_df.groupby([genre, 'release_year']).mean()['popularity'][1]
    popularity_genre_year.loc[genre] = avg_popularity

popularity_genre_year = popularity_genre_year.fillna(value=0)
popularity_genre_year
```

Created a new dataframe to contain average popularity of each genre over the years. Next  I'll use various plot to understand the trends.

```python
# plot the distribution of average popularity of each genre over the years
popularity_genre_year.transpose().plot(figsize=(20, 8))
plt.show()
```

```python
# distribution of popularity of each genre over the years
fig, axes = plt.subplots(nrows=10, ncols=2, figsize=(20, 50))
popularity_genre_year.transpose().plot(kind='bar', ax = axes, subplots=True)
plt.show()
```

### Investigate properties of high revenue 

Now, I'll investigate how the revenue varies with other parameters. First, I'll investigate how average revenue varies acroos genre.

```python
# The distribution of popularity over each genre
distribution_over_genre('revenue')
```

The plot indicates that Adventure, Fantasy, Animation, Science Fiction, Family and Action movies tend to genrate higher revenue compared to others.

```python
years = sorted(tmdb_df['release_year'].unique(), reverse=True)

# utility function plot distribution of data over years
def distribution_var_years(var):
    '''
    This function takes a variable as input and plots the distribution 
    of that variable over years
    '''
    distribution_years = pd.Series(index=years)
    for year in years:
        distribution_years[year] = tmdb_df.groupby('release_year').mean()[var][year]
    title = 'Distribution of ' + var + ' over years'
    distribution_years.plot(kind='bar', figsize=(20, 5), title = title)
    plt.show()
```

```python
# plot distribution of revenue over years
distribution_var_years('revenue')

# plot distribution of revenue_adj over years
distribution_var_years('revenue_adj')
```

The first plot above shows revenue acroos the years and it seems that after 90s the average revenue seeems to a lot more than previous years. But the second plot above shows the complete opposite. This is actually a very good insight. Becuase the earlier plot was on revenue field, and the later one is on revenue_adj which is more appropriate as it combines the inflation information over the years. So, in my opinion, using revenue_adj is more appropriate measure of revenue. I'll use revenue_adj in future plots.

```python
# utility function to create scatter plot of var1 and var2
def plot_scatter(x_var, y_var):
    fig = plt.figure(figsize=(10,4))
    plt.scatter(np.array(tmdb_df[x_var]), np.array(tmdb_df[y_var]))
    plt.title('Scatter plot of ' + x_var + ' vs ' + y_var)
    plt.xlabel(x_var)
    plt.ylabel(y_var)
    plt.show()
```
